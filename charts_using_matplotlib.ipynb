```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df0 = df[df["target"] == 0]
df1 = df[df["target"] == 1]
df2 = df[df["target"] == 2]
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='.')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'],color="red",marker='*')
```

```python
sns.boxplot(x='target', y='sepal length (cm)', data=df)
plt.title('Sepal Length by Target')
plt.xlabel('Target')
plt.ylabel('Sepal Length')
plt.show()
```

```python
avg_sepal_length = df.groupby('target')['sepal length (cm)'].mean()
```

```python
plt.bar(avg_sepal_length.index, avg_sepal_length)
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length')
plt.xticks(avg_sepal_length.index, ['Setosa', 'Versicolor', 'Virginica'])
plt.show()
```
